# 01 — Inspect a messy dataset

First-pass helpers for any new DataFrame: per-column overview, dtype counts, missingness, categorical summary, grouped numeric summary, and a one-shot inspector.

```python
import pandas as pd

from geo_daily_tools.sample_data import messy_sensor_records
from geo_daily_tools.inspection import (
    column_overview,
    dtype_summary,
    missingness_report,
    categorical_summary,
    grouped_numeric_summary,
    quick_inspect,
)

df = pd.DataFrame(messy_sensor_records())
df
```

## Per-column overview — dtype, missing, unique, sample value

```python
column_overview(df)
```

## Dtype distribution

```python
dtype_summary(df)
```

## Missingness ranking

```python
missingness_report(df)
```

## Categoricals at a glance

```python
categorical_summary(df)
```

## Numeric summary by group

```python
grouped_numeric_summary(df, group_cols='sensor_type', value_cols='reading')
```

## One-shot inspection (prints everything)

```python
_ = quick_inspect(df, name='messy_sensor_records')
```
